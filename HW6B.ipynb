```python
url = 'https://dl.dropboxusercontent.com/u/48599099/Stat133/data/ALL_GRADES.csv'
grades = pd.read_csv(url) 

total_grades = 0.3*grades['F'] + 0.25*grades['M1'] + 0.25*grades['M2'] + 0.20*grades['THW'] 
grades['TG'] = total_grades

total_grades.hist()

assigngrades = total_grades.copy()
assigngrades.sort()

grades.head(20)
assigngrades.head(20)

grade_A = DataFrame(assigngrades[255*0.7:]).index
grade_B = DataFrame(assigngrades[255*0.4:255*0.7]).index
grade_C = DataFrame(assigngrades[255*0.2:255*0.4]).index
grade_D = DataFrame(assigngrades[255*0.1:255*0.2]).index
grade_F = DataFrame(assigngrades[:255*0.1]).index

lettergrade={}

for item in grade_A:
    lettergrade[item] = 'A'
for item in grade_B:
    lettergrade[item] = 'B'
for item in grade_C:
    lettergrade[item] = 'C'    
for item in grade_D:
    lettergrade[item] = 'D'
for item in grade_F:
    lettergrade[item] = 'F'    
    
lettergradelist = Series(lettergrade)
grades.loc[:,'LG'] = lettergradelist
print grades.head(20)
```
