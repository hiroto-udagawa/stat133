###Exercice 1. *Fill lists with function values.*
Write a Python functin representing the mathatematical function
$$h(x)=\frac1{\sqrt{2\pi}}e^{-\frac12x^2}.$$

```python
def h(x):
    N = np.sqrt(2*np.pi)
    A = 0.5*x**2
    return (1/N)*np.exp(-A)
```

###Exercise 4. * A wave class with plotting capabilities. *
The collection of functions

$$f(x,t; a)=e^{−a(x−3t)^2} \sin(3\pi(x−t))$$

describes a wave localized in space ($x$ variable) at time $t$, where $a$ is a parameter.  

Write a class <code>Wave</code> representing this family of functions satisfying the following specifications:

The constructor should be of the form

    def __init__(self, a=1, interval=[-4, 4], pt_num=40):
where $a$ is the parameter,  <code>interval</code> is a list consisting of the lower-bound and the upper-bound for the $x$-variable, and <code>pt_num</code> is the number of points to be sampled for the $x$ variable when plotting. The constructor should store the values passed to it in the class attributes:

        self.a = a
        self.x0 = interval[0]
        self.x1 = interval[1]
        self.pt_num = pt_num

You should implement the following methods:

    def compute(self, x,t):
that should return the value $f(x,t;a)$ with parameter $a$ having the value stored in the attribute <code>self.a</code>.
    
    def x_values(self):
that should return a numpy array of <code>self.pt_num</code> points between <code>self.x0</code> and 
<code>self.x1</code>

    def plot(self, time):
that should draw four plots of the function $f$ arranged in a ROW (use the <code>subplots</code> method) at different times (equally spaced between $0$ and <code>time</code>. The $x$-axis should be labelled 'x' and the $y$-axis should be labelled 'y'. The title of each of the plot should indicate the time at which the wave is plotted. 
 

```python
class Wave:    
    def __init__(self, a=1, interval=[-4, 4], pt_num=40):
        self.a = a
        self.x0 = interval[0]
        self.x1 = interval[1]
        self.pt_num = pt_num
        
    def compute(self, x,t):      
        return np.exp(-self.a*(x - 3*t)**2)*np.sin(2*np.pi*(x - t))
    
    def x_values(self):
        return np.linspace(self.x0, self.x1, self.pt_num)
    
    def plot(self, time):
        x = self.x_values()
        nrows = 1
        ncols = 4
        nframes = nrows*ncols

        fig, axes = plt.subplots(nrows=1, ncols=nframes, figsize=(12,3))
        times  = np.linspace(0, time, nframes)

        i = 0
        for t in times:
            y = self.compute(x, t)
            axes[i].plot(x, y)
            axes[i].set_xlabel('x')
            axes[i].set_ylabel('y')
            axes[i].set_title('Wave at t = %4.4f' % t)
            i += 1
        fig.tight_layout()
```

Test the class by executing the following cell (you can play with the interval and the parameter $a$):

```python
a = 2
interval = [-1, 1]
time_span = 1

F = Wave(a, interval)
F.plot(time_span)
```
